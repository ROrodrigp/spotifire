```python
import pandas as pd
import boto3
from io import BytesIO
```

```python
# Configurar S3
# Definir la ruta exacta al archivo snappy.parquet
bucket_name = "itam-analytics-ragp"
parquet_key = "spotifire/processed/likes/user_12137259902.parquet/part-00000-650e9fc6-e797-4819-8b18-e54b0fe1e024-c000.snappy.parquet"
s3_client = boto3.client('s3')
```

```python
# Descargar el archivo directamente
print(f"Descargando archivo: s3://{bucket_name}/{parquet_key}")
response = s3_client.get_object(Bucket=bucket_name, Key=parquet_key)
parquet_data = BytesIO(response['Body'].read())
```

```python
#  Leer el parquet con pandas
df = pd.read_parquet(parquet_data)
```
